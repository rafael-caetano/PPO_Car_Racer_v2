# PPO and SAC implementations for CarRacubf-v2

## PPO Continuous

```python
import gymnasium as gym
from IPython.display import Video

from src.utils import FrameStack, plot_training_progress, record_video
from src.train import train_ppo

# Train the agent
ppo_agent, rewards, actor_losses, critic_losses, entropies, iterations = train_ppo(num_episodes=2000, max_steps=1000, update_interval=2048, plot_interval=10)

# Save the final plot
plot_training_progress(rewards, actor_losses, critic_losses, entropies, iterations, save=True)

# Create a new environment for video recording
eval_env = gym.make('CarRacing-v2', render_mode="rgb_array")
eval_env = FrameStack(eval_env)

# Record a video of the trained agent
final_reward = record_video(eval_env, ppo_agent)
print(f"Final episode reward: {final_reward:.2f}")

# Display the video
Video("outputs/trained_ppo_continuous_agent_video.mp4")
```
